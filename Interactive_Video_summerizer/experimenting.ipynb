```python
import whisper

# Load the Whisper model
model = whisper.load_model("medium")  # or "turbo" if you prefer

# Transcribe the audio file
result = model.transcribe(mainPath+'/data/Youtube_video_audio.wav')

# Extract and print sentence-level timestamps and transcription
for segment in result['segments']:
    start = segment['start']
    end = segment['end']
    text = segment['text'].strip()  # Clean up any extra whitespace
    print(f"[{start:.3f}s --> {end:.3f}s] {text}")
```

```python
import torch
from transformers import AutoProcessor, AutoModelForCTC
from pydub import AudioSegment

# Load the pre-trained model and processor for Vietnamese speech-to-text
processor = AutoProcessor.from_pretrained("leduytan93/Fine-Tune-XLSR-Wav2Vec2-Speech2Text-Vietnamese")
model = AutoModelForCTC.from_pretrained("leduytan93/Fine-Tune-XLSR-Wav2Vec2-Speech2Text-Vietnamese")

# Load your audio file (ensure it's a .wav file)
audio_path = mainPath+'/data/Youtube_video_audio.wav' # Replace with your audio file path
audio = AudioSegment.from_file(audio_path, format="wav")

# Convert audio to raw waveform and sample rate (ensure it's mono)
samples = audio.set_channels(1).get_array_of_samples()
samples = torch.tensor(samples).float() / 32768.0  # Normalize the waveform

# The model expects input with a sample rate of 16kHz
sample_rate = 16000
if audio.frame_rate != sample_rate:
    audio = audio.set_frame_rate(sample_rate)
    samples = audio.set_frame_rate(sample_rate).get_array_of_samples()  # Get samples again after frame rate change
    samples = torch.tensor(samples).float() / 32768.0  # Normalize again

# Preprocess the audio
inputs = processor(samples.numpy(), sampling_rate=sample_rate, return_tensors="pt", padding=True)

# Perform inference
with torch.no_grad():
    logits = model(inputs.input_values).logits

# Decode the predicted ids to text
predicted_ids = torch.argmax(logits, dim=-1)
transcription = processor.batch_decode(predicted_ids)

# Print the transcription
print("Transcription:", transcription[0])
```

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lex_rank import LexRankSummarizer

# Load your transcript text
transcript_text = result['text']

# Initialize parser and summarizer
parser = PlaintextParser.from_string(transcript_text, Tokenizer("english"))
summarizer = LexRankSummarizer()

# Summarize to 5 key sentences (adjust as needed)
summary = summarizer(parser.document, 20)
for sentence in summary:
    print(sentence)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Load the model and tokenizer
sum_tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
sum_model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")
```

```python

# Your transcript text
transcript = result['text']
# Tokenize the input text and set max length (you can split long text if needed)
inputs = sum_tokenizer.encode("summarize: " + transcript, return_tensors="pt", max_length=1024, truncation=True)

# Generate the summary
summary_ids = sum_model.generate(inputs, max_length=10000, min_length=100, length_penalty=2.0, num_beams=4, early_stopping=True)

# Decode the summary
summary = sum_tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Display the summary
print("Summary:", summary)
summery=[]
for sent in summary.split('.'):
  summery.append(sent)
```

```python
from sentence_transformers import SentenceTransformer, util
import numpy as np

# Initialize the model for embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')  # Lightweight model, change as needed

# Example transcript with timestamps
transcript = original_transcript

# Summarized bullet points (from a summarizer like BART)
summary = summery

# Step 1: Generate embeddings for transcript and summary
transcript_sentences = [sent[2] for sent in transcript]
transcript_embeddings = model.encode(transcript_sentences, convert_to_tensor=True)
summary_embeddings = model.encode(summary, convert_to_tensor=True)

# Step 2: Calculate similarity and find the closest match for each summary point
for i, sum_emb in enumerate(summary_embeddings):
    # Compute cosine similarities between the summary and all transcript sentences
    cosine_scores = util.cos_sim(sum_emb, transcript_embeddings)[0]

    # Get the index of the transcript sentence with the highest similarity score
    most_similar_idx = np.argmax(cosine_scores.cpu().numpy())

    # Extract timestamp of the most similar sentence
    timestamp = transcript[most_similar_idx][:2]
    print(f"Summary Point {i+1}: '{summary[i]}'")
    print(f"   Closest match in transcript (Timestamp: {timestamp[0]}s - {timestamp[1]}s): '{transcript[most_similar_idx][2]}'")
    print()
```
